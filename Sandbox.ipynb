```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('completo.csv')
df
```

```python
df["Data"] = pd.to_datetime(df["Data"], infer_datetime_format=True)  
df.groupby(["Cidade",pd.Grouper(key='Data', freq='W')]).mean()
#df["Data"].dt.to_period("W")
```

```python
alt.Chart("https://raw.githubusercontent.com/tiagodavi70/altair_tutorial/master/completo.csv").mark_line().encode(
    alt.X('month(Data):T'),
    alt.Y('average(VENTO | VELOCIDADE HORARIA (m/s)):Q')
)
```
